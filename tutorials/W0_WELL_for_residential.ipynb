**Diving into WELL Standard**

The [WELL Building Standard™](https://v2.wellcertified.com/en/wellv2/overview) version 2 (WELL v2™) is a vehicle for buildings and organizations to deliver more thoughtful and intentional spaces that enhance ***human health and well-being***. WELL v2 includes a set of strategies—backed by the latest scientific research—that aim to advance human health through design interventions and operational protocols and policies and foster a culture of health and well-being. Built upon the pioneering foundation of the first version of the WELL Building Standard (WELL v1), WELL v2 draws expertise from a diverse community of WELL users, practitioners, public health professionals and building scientists around the world.

There are ***ten concepts*** in WELL v2:

<img src="https://drive.google.com/uc?export=view&id=1Ek-vhW23h4OuNrbxElfvlOlS9NlicSpd" alt="isometric" class="center" width="600px">

10 well concepts for health and well-being of building occupants - source: [WELL certified](https://www.wellcertified.com)

Each concept consists of ***features*** with distinct health intents. Features are either *preconditions* or *optimizations*.

Let's go through some WELL guidelines for ***Dwelling Units and Guest Rooms***. The source of the following concepts are a mixture of WELL V2 and WELL for Residential.

At least one functional building ***entrance*** opens to a ***pedestrian*** network (i.e., streets where pedestrians travel featuring, at a minimum, sidewalks) and meets at least one of the following requirements:


*   The project's address achieves a minimum *Walk Score*® of 70
*   Opens onto a street with restricted *vehicular*  traffic

1.   At least eight existing use types (as defined in Appendix V1) are present.

```python
#importing the libraries
import drawsvg as draw
import os
from PIL import Image
from IPython.display import display, SVG

#resizing the image to 550x300 pixels
image_path = "/content/Stack.png"  # Change this to your image file path
resized_image_path = "resized_example.png"

img = Image.open(image_path)
img = img.resize((550, 300))  # Resize to 500x300
img.save(resized_image_path)  # Save resized image

#creating the SVG with embedded image
image_width, image_height = 550, 300
d = draw.Drawing(image_width, image_height)

#embedding the resized image
image_element = draw.Image(0, 0, image_width, image_height, resized_image_path)
d.append(image_element)

#drawing a circle on top of the image
circle = draw.Circle(400, 150, 20, fill='red', stroke='black', stroke_width=1)  # (cx, cy, radius)
d.append(circle)

#adding a text on top of the image
d.append(draw.Text('The Stack', 20, 300, 100, fill='blue'))


#save and display the SVG
d.save_svg("output.svg")

!convert output.svg output.png #is a shell command that runs in Google Colab. It uses ImageMagick's convert tool to convert the SVG file (output.svg) into a PNG file (output.png).
from IPython.display import Image
display(Image("output.png"))  #display the PNG
```

Here you can choose among these annotation options to add them to code cell above. It's practical if you create a new code cell and copy the parts that are needed from both cells.

```python
d = draw.Drawing(200, 100, origin='center')

# Draw an irregular polygon
d.append(draw.Lines(-80, 45,
                     70, 49,
                     95, -49,
                    -90, -40,
                    close=False,
            fill='pink',
            stroke='black'))

# Draw a rectangle
r = draw.Rectangle(-80, -50, 40, 50, fill='#1248ff')
r.append_title("Our first rectangle")  # Add a tooltip
d.append(r)

# Draw a circle
d.append(draw.Circle(-40, 10, 30,
        fill='purple', stroke_width=2, stroke='black'))

# Draw an arbitrary path (a triangle in this case)
p = draw.Path(stroke_width=2, stroke='lime', fill='black', fill_opacity=0.2)
p.M(-10, -20)  # Start path at point (-10, -20)
p.C(30, 10, 30, -50, 70, -20)  # Draw a curve to (70, -20)
d.append(p)

# Draw text
d.append(draw.Text('Basic text', 8, -10, -35, fill='blue'))  # 8pt text at (-10, -35)
d.append(draw.Text('Path text', 8, path=p, text_anchor='start', line_height=1))
d.append(draw.Text(['Multi-line', 'text'], 8, path=p, text_anchor='end', center=True))

# Draw multiple circular arcs
d.append(draw.ArcLine(60, 20, 20, 60, 270,
        stroke='red', stroke_width=5, fill='red', fill_opacity=0.2))
d.append(draw.Arc(60, 20, 20, 90, -60, cw=True,
        stroke='green', stroke_width=3, fill='none'))
d.append(draw.Arc(60, 20, 20, -60, 90, cw=False,
        stroke='blue', stroke_width=1, fill='black', fill_opacity=0.3))

# Draw arrows
arrow = draw.Marker(-0.1, -0.51, 0.9, 0.5, scale=4, orient='auto')
arrow.append(draw.Lines(-0.1, 0.5, -0.1, -0.5, 0.9, 0, fill='red', close=True))
p = draw.Path(stroke='red', stroke_width=2, fill='none',
        marker_end=arrow)  # Add an arrow to the end of a path
p.M(20, 40).L(20, 27).L(0, 20)  # Chain multiple path commands
d.append(p)
d.append(draw.Line(30, 20, 0, 10,
        stroke='red', stroke_width=2, fill='none',
        marker_end=arrow))  # Add an arrow to the end of a line

d.set_pixel_scale(2)  # Set number of pixels per geometry unit
#d.set_render_size(400, 200)  # Alternative to set_pixel_scale
d.save_svg('example.svg')
d.save_png('example.png')

# Display in Jupyter notebook
#d.rasterize()  # Display as PNG
d  # Display as SVG
```
